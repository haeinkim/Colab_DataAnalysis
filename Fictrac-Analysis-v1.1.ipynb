```python
data_header = ['_', '_', '_', '_', '_',
               'dRotVx', 'dRotVy', 'dRotVz',
               '_', '_', '_', '_', '_', '_',
               'x', 'y', 'h',
               '_', '_',
               '_',
               '_',
               'timestamp',
               'seq_counter']

frame_rate = 30
# sample_rate = 0.2
sample_rate = 1
ball_radius = 10
set_time = 180

appended_data = []

for i, d in enumerate(base_dir):
    fnames = glob.glob(d)
    print(fnames)

    for j, f in enumerate(fnames):

        df = pd.read_csv(f, header=0)
        df.columns = data_header
        df.drop(['_'], axis=1, inplace=True)
        dt = df['timestamp'].diff() / 1000
        df['time'] = dt.cumsum()
        df[['Vx', 'Vy', 'Vz']] = df[['dRotVx', 'dRotVy', 'dRotVz']].div(dt, axis=0)
        df[['x', 'y', 'Vx', 'Vy']] = df[['x', 'y', 'Vx', 'Vy']] * ball_radius
        df['Vz'] = np.rad2deg(df['Vz'].values)
        df[['Sx', 'Sy', 'Sz']] = df[['Vx', 'Vy', 'Vz']].abs()

        # grouping and averaging data in 200ms time window 
        df = df[df['time'] <= set_time + 1]
        last = df['time'].values[-1]
        rng = np.arange(0, last+sample_rate, sample_rate)
        ds = df.groupby(pd.cut(df['time'], rng)).mean()
        ds.interpolate(method='linear', limit_direction='both', axis=0, inplace=True) # filling gap b/w data points
        ds['group'] = i
        ds['fly'] = j
        ds.drop(['h', 'timestamp', 'seq_counter'], axis=1, inplace=True)
        ds.reset_index(drop=True, inplace=True)
        ds['time'] = sample_rate * ds.index
        ds['x'] = ds['x'] - ds['x'][0]
        ds['y'] = ds['y'] - ds['y'][0]
        appended_data.append(ds)
        # df.fillna(0, inplace=True)

df = pd.concat(appended_data)
df.tail()
```

```python
from scipy.spatial import distance
num_fly = len(df['fly #'].unique())
print(num_fly)
print(df['Category'].unique())
event = ['before', 'event','after']

i=1

appended_data = []
data = df
for t, e in enumerate(event):
    for n in range(num_fly):
        data = df[(df['time [s]']>=t*60) & (df['time [s]']<(t+1)*60)]
        sample = data[(data['Category'] == '+ ATR') & (data['fly #'] == n+1)]
        coords = sample[['x', 'y']]
        d_mat = distance.cdist(coords, coords, metric='euclidean')
        if t*60 == 0:
            dis_cumsum = np.flip(d_mat.diagonal(-1)).cumsum()
            idx = pd.Index(dis_cumsum).get_loc(150, 'nearest')
            # print(t*60)
        else:
            dis_cumsum = d_mat.diagonal(-1).cumsum()
            idx = pd.Index(dis_cumsum).get_loc(150, 'nearest')
        # print(idx)
        paths = (sample[:idx][['x','y']]-sample.iloc[0][['x','y']])
        # idx = pd.Index(pd.Series(np.flip(d_mat.diagonal(-1)).cumsum())).get_loc(150, 'bfill')
        # paths = (sample[-idx:][['x','y']]-sample.iloc[-1][['x','y']])
        paths['Xmean'] = paths['x'].mean()
        paths['Ymean'] = paths['y'].mean()
        paths['fly #'] = n+1
        paths['event'] = e
        _x = paths['Xmean'].unique()
        _y = paths['Ymean'].unique()
        paths['dist'] = distance.euclidean((_x, _y), (0,0))
        appended_data.append(paths)
output = pd.concat(appended_data).reset_index(drop=True)
output.iloc[:,:2] = output.iloc[:,:2].apply(pd.to_numeric)
output.tail()
```
